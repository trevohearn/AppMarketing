# Imports

```python
from bs4 import BeautifulSoup
import requests

URL = 'https://www.hertzcarsales.com/massapequa/used-cars.htm'
AGENT = {"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.87 Safari/537.36'}
COOKIES = {"cookie": None}

page = requests.get(URL, headers=AGENT, cookies=COOKIES)
bs = BeautifulSoup(page.content, 'html.parser')
```
