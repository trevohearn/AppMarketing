```python
#$##
def dropdollarsign(nums):
    cleaned = []
    for x in nums:
        if (x[0] == '$'):
            cleaned.append(int(x[1:]))
        else:
            cleaned.append(int(x))
    return cleaned
price_list_ints = dropdollarsign(price_list)
```
